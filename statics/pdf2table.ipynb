```python
import tabula

# PDF 파일 경로 설정
file_path = '/content/table01.pdf'

# PDF 파일에서 테이블 추출
df_list = tabula.read_pdf(file_path, pages='all', multiple_tables=True)

# 추출된 테이블을 DataFrame으로 변환
for i, df in enumerate(df_list):
    print(f"Table {i+1}:")
    print(df)
```

```python
import tabula

# Attempt to read the table from the newly provided PDF file path
file_path = '/content/table02.pdf'

# Try to extract tables from the PDF
try:
    df_list = tabula.read_pdf(file_path, pages='all', multiple_tables=True)
    df = df_list[0] if df_list else None
except Exception as e:
    df = f"Failed to read the table: {str(e)}"

df
```
